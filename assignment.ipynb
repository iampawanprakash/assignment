```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("AgeGroupDetails.csv")
plt.plot( df['AgeGroup'],df['TotalCases'],marker="o",color="red",label="cases")
plt.plot(df["AgeGroup"],df["Gender"],marker="o",color="red")
plt.bar(df["AgeGroup"],df["TotalCases"],label="male")
plt.bar(df["AgeGroup"],df["Gender"],label="female")
plt.legend()
plt.xlabel("Age groups")
plt.ylabel("number of cases")
plt.title("number of cases according to age groups")
plt.show()

```
